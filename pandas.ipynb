```python
weather = pd.read_csv('data/weather.csv')
weather.head()
```

```python
weather_melt = pd.melt(weather, 
                       id_vars=['id', 'year', 'month', 'element'],
                      var_name='day', value_name='temp')
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(index=['id', 'year','month','day'],
                        columns='element', values='temp')
weather_tidy.head()
```

```python
weather_tidy = weather_tidy.reset_index()
weather_tidy
```

```python
weather_tidy.info()
```

```python
weather_tidy.loc[
    (weather_tidy.year==2010) &
    (weather_tidy.month==2) &
    (weather_tidy.day=='d1')]
```

```python
weather_tidy.loc[
    (weather_tidy.year==2010) &
    (weather_tidy.month==12) &
    (weather_tidy.day=='d6'), ['tmax', 'tmin']]
```

```python
weather_tidy.head()
```

```python
def mean_func(x):
    return sum(x) / len(x)
```

```python
def avg(a, b):
    return (a + b)/2
```

```python
weather_tidy[['tmin', 'tmax']].apply(mean_func, axis=1)
```
