```python
import pandas as pd
```

```python
pop = pd.read_csv('world-bank-data-1960-to-2016/country_population.csv')
pop.head()
```

```python
fer = pd.read_csv('world-bank-data-1960-to-2016/fertility_rate.csv')
fer.head()
```

```python
life = pd.read_csv('world-bank-data-1960-to-2016/life_expectancy.csv')
life.head()
```

```python
population = pd.melt(pop, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='pop')
population.head()
```

```python
fertility = pd.melt(fer, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='fer_rate')
fertility.head()
```

```python
life_exp = pd.melt(life, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='life_exp')
life_exp.head()
```

```python
s1 = fertility.iloc[:,5:6].copy()
s1.head()
```

```python
s2 = life_exp.iloc[:,5:6].copy()
s2.head()
```

```python
new_table = pd.concat([population, s1], axis=1)
new_table.head()
```

```python
final_table = pd.concat([new_table, s2], axis=1)
final_table.head()
```

```python
del final_table['Indicator Name']
```

```python
del final_table['Indicator Code']
```

```python
final_table.head()
```

```python
final_table.groupby('Year')[['Country Name', 'Country Code', 'pop', 'fer_rate', 'life_exp']].mean().reset_index().head()
```

```python
final_table.head()
```

```python
final_table.to_csv('outputs/final_table.csv', sep=',')
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
india = final_table.loc[final_table['Country Name'] == 'India']
china = final_table.loc[final_table['Country Name'] == 'China']
```

```python
china.head()
```

```python
x = range(1960, 2017)
y_india = india.iloc[:,3:4]
y_china = china.iloc[:, 3:4]

plt.title('China vs India')
plt.ylabel('population(billions)')
plt.xlabel('year')

plt.plot(x, y_india, label='India')
plt.plot(x, y_china, label='China')

plt.legend()

plt.show()
```

```python
x = range(1960, 2017)
y_india = india.iloc[:,4:5]
y_china = china.iloc[:, 4:5]

plt.title('China vs India')
plt.ylabel('fertility rate(births per woman)')
plt.xlabel('year')

plt.plot(x, y_india, label='India')
plt.plot(x, y_china, label='China')

plt.legend()

plt.show()
```

```python
x = range(1960, 2017)
y_india = india.iloc[:,5:6]
y_china = china.iloc[:, 5:6]

plt.title('China vs India')
plt.ylabel('life expectancy')
plt.xlabel('year')

plt.plot(x, y_india, label='India')
plt.plot(x, y_china, label='China')

plt.legend()

plt.show()
```
